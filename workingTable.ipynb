```python
import pandas as pd
from IPython.display import display
import numpy as np


# To import the Excel Workbook
xls_file = pd.ExcelFile('HC_Jenny.xlsx')

# To select the Excel Worksheet of your interest
table = xls_file.parse('raw_code')
display(table)
```

```python
# To select filter out data that less than 15 from the column1
df = table[table.Taqman_IU > 15]

# To select filter out data that less than 15 from the column2
df = df[df.cobas_IU > 15]

# To count and verify total datapoint
df.count()
```

```python
import numpy as np

# To assign the values of column1 to x-axis
x = df['Taqman_Log']

# To assign the values of column2 to y-axis
y = df['cobas_Log']

# To get the linear regression and print the best fit line formula (Minimal sum of square method). 
regression = np.polyfit(x, y, 1)
print('y =', regression[0], 'x +', regression[1])

# To get the correlation, r, for all the data.
display(df.corr())

# To get the correlation, r2, for the target of interest, column1 vs column2.
print('r2 =',(0.984582)**2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# A scatter plots for x vs y values
plt.figure(figsize=(15, 10))
sns.regplot(x=x, y=y, truncate=True)
plt.title('Comparison between cobas® HCV 6800 and COBAS® AmpliPrep/COBAS® TaqMan® HCV Test')
plt.xlabel('AmpliPrep/TaqMan (Log IU/mL)')
plt.ylabel('cobas HCV 6800 (Log IU/mL)')
plt.text(6, 3.8, r'y = 1.0235x - 0.0270')
plt.text(6, 3.5, r'r2 = 0.9694')
plt.show()
```

```python
# To define function to generate bland_altman_plot
def bland_altman_plot(data1, data2, *args, **kwargs):
    data1     = np.asarray(data1)
    data2     = np.asarray(data2)
    mean      = np.mean([data1, data2], axis=0)
    diff      = data1 - data2                   # Difference between data1 and data2
    md        = np.mean(diff)                   # Mean of the difference
    sd        = np.std(diff, axis=0)            # Standard deviation of the difference
    
    # to print mean, mean+-2sd
    print(md)
    print(md+ 1.96*sd)
    print(md- 1.96*sd)
    
    plt.figure(figsize=(15, 10))
    plt.scatter(mean, diff, *args, **kwargs)
    plt.axhline(md,           color='gray', linestyle='--')
    plt.axhline(md + 1.96*sd, color='gray', linestyle='--')
    plt.axhline(md - 1.96*sd, color='gray', linestyle='--')

bland_altman_plot(y, x)
plt.title('Bland-Altman plot comparing the difference between the cobas® HCV 6800 and COBAS® AmpliPrep/COBAS® TaqMan® HCV Test')
plt.xlabel('Average of 2 measures')
plt.ylabel('Difference between 2 measures')
plt.text(1.2, 0.57, r'Upper limit of 95% CI (0.6363)')
plt.text(1.2, -0.5, r'Lower limit of 95% CI (-0.431)')
plt.text(1.2, 0.05, r'Mean difference (0.1025)')
plt.show()
```
